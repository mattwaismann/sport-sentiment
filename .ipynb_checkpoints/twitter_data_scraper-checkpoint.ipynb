```python
url = 'https://www.dataforprogress.org/blog/2018/11/19/identifying-and-estimating-the-ideologies-of-twitter-pundits'
res = requests.get(url)
soup = BeautifulSoup(res.text)
data_table = soup.select("td")
process = [table.text for table in data_table]
process.pop(0)
```

```python
process = np.array(process)
#every 5th element from the index 0
handles = process[0::5]
#every 5th element from the index 1
politics = process[1::5]
data = pd.DataFrame({'screen_name':handles, 'political_leaning':politics})
data.to_csv("twitter_politics.csv", header = True,  index = False)
```
